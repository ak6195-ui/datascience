# PYTHON OS MODULE 
 1) os.getcwd(): get the current working directory 
 2) os.mkdir('path'):  create a new directory 
 3) os.chdir('path'): change the current working directory 
 4) os.listdir('path'):list the contents of a directory 
 5) os.rename('old name ','new name '):renames a file for directory 
 6) os.remove('file to delete'):removes a file
 7) os.rmdir('dir to delete'):remove directory 
 8) os.path.dirname('path'):returns the name of the directory containing path name path
 9) os.path.exists('path'):checks if the path already exists
