# lambda function
lambda arguments:expression


```python
def square(base,power):
    return base**power
square(2,3)
```

```python
power_up=lambda base,power:base**power
type(power_up)
```

```python
power_up(3,3)
```

```python
area=lambda l,b,h:2*((l*b)+(b*h)+(l*h))
```

```python
area(3,3,3)
```
