```python
def cal_area(shape,**kwargs):
    if shape=="square":
        area=kwargs.get("length",0)**2
        print(area)
    return
```

```python
from math import sqrt
def stdd(value:list):
    N=len(value)
    mean=(sum(value)/N)
    temp=sum([(x -mean)**2 for x in value])
    std=sqrt(temp/N)
    return std
    
      
    
```

```python
stdd([1,2,3,4,5,6,7,8,9])
```

```python
import random
def gusss(*val):
    num=random.randint(1,9)
    print(num)
    i=0
    while i<=10:
        if val[i]==num:
            print("CONGRALUTION")
            break
        elif(val[i]<=num):
            print("IS LESS")
            i+=1
        elif(val[i]>=num):
            print("IS MORE")
            i+=1
    return
```

```python
gusss(1,2,3,4,5,6,7,8,9)
```
