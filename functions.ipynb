# enumerate()

```python
names=["abo","jaja","kaka","lala"]
list(enumerate(names))
```

```python
for (i,name) in enumerate(names ,start=1):
    print(f"{name}:roll no {i}")
```

# zip() 
le iterable jodxa 

```python
nam=["abi","khadka","shyam"]
score=[1,2,3]
list(zip(nam,score)
```

```python
for (n,s) in zip(nam,score):
    print(f"name\t{n}:score is {s}")
```

```python
set1=set("aeiou")
list(zip(nam,score,set1))
```

# Filter()
yesle condition milne wala matra firta dinxa

```python
def even(num):
    return num%2==0
```

```python
nk=[1,2,3,4]
filt=filter(even,nk)
print(list(filt))
```
