# list comprehension 

```python
# output=[<expression> for items in iterables if <condition>]
y=[i**2 for i in range(1,10)]
y
```

```python
ev=[]
for i in range(1,20):
    if i%2==0:
        ev.append(i**2)
    else :
        continue
print(ev)
```

```python
rr=[i**2 for i in range (1,20) if i%2==0]
rr
```
