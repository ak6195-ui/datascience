# Encapsulation in Python
1.private(_2_) \
2.protected(_1)

```python
class Example:
    def __init__(self):
        self.public=1
        self.__private=2
        self._protected=3
    def public_met(self):
        print(self.__private)
        print("public")
    def __private_m(self):
        print("private method")
    def _protect_m(self):
        print("protected")
```

```python
obj=Example()
```

```python
obj.public
```

```python
obj.public_met()
```

```python
obj._protected
```

```python
#obj.__private MILDAINA
```

```python
class subclass(Example):
    def get_public(self):
        print(self.public)

    def get_protected(self):
        print(self._protected)

    def get_private(self):
        print(self.__private)
        
```

```python
sub=subclass()
```

```python
sub.get_public()
```

```python
sub.get_protected()
```
