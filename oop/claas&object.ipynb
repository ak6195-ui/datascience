```python
class Calculator:
    def __init__(self,num1,num2):
        self.num1=num1 #self.value1=num1
        self.num2=num2
        
    def sum(self,num3):
        return self.num1+self.num2 +num3
    def mul(self):
        return self.num1*self.num2
```

```python
calc=Calculator(10,4)
calc.sum(2)
```

```python
calc.mul()
```

# Class Methods

```python
class Ola:
    count=0
    def __init__(self,name:str,address="ktm"):
        self.name=name
        self.address=address
        
        Ola.count += 1
    def get_details(self): #instance method
        print(f"i am {self.name} and i stay at {self.address}")
    @classmethod
    def get_count(cls):#class method
        return cls.count
    @staticmethod
    def create_name(first,last):
        return f"{last},{first}"
    
    
        
        
```

```python
st1=Ola("abi")
```

```python
st1.name
```

```python
st1.get_details()
```

```python
st1.get_count()
```

```python
st1.create_name("abi","khadka")
```

```python
class Stu2:
    def __init__(self,name,marks:list[int]):
        self.name=name
        self.marks=marks
    def tot(self):
        return sum(self.marks)
    def average(self):
        return self.tot()/len(self.marks)
    def display(self):
        print(f"name of student is: {self.name}")
        print(f"the average marks is: {self.average()}")
        print(f"the total marks obtained is : {self.tot()}")
        
```

```python
pr=Stu2("abi khadka",[90,90,90])
```
