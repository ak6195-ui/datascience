# DUNDERS= __

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"{self.name},{self.age} "
    def __repr__(self):
        return f"{self.name},{self.age} "
    def __len__(self):
        return self.age
    def __call__(self,gender="male"):
        return f"{self.name} is a {gender}"
```

```python
person=Person("abi",18)
```

```python
class Point :
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __str__(self):
        return f"point({self.a},{self.b})"
        
    def __add__(self,other):
        a= self.a + other.a
        b= self.b + other.b
        return Point(a,b)
    def __mul__(self,other):
        a= self.a * other.a
        b= self.b * other.b
        return Point(a,b)
    # __eq__ le chai equal check garxa 
    def __eq__(self,other):
        return self.a==other.a and self.b==other.b
        
```

```python
a=Point(2,3)
b=Point(4,5)
```

```python
print(a)
```
