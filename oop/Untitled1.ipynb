# Single Inherintance

```python
class Animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def into(self):
        print(f"my name is: {self.name}")
    
```

```python
class Cat(Animal):
    def __init__(self,name,age,color):
        super().__init__(name,age)
        self.color=color
    def details(self):
        print(f"{self.name},{self.age}")
    def sound(self):
        print("meoww")
```

```python
cat=Cat("la",10,"black")
```

```python
cat.into()
cat.sound()
```

```python
cat.details()
```

```python
class La(Cat):
    def __init__(self,name,age,color,anni):
        super().__init__(name,age,color)
        self.anni=anni
    def pr(self):
        print(self.name)
        print(self.age)
        print(self.color)
        print(self.anni)
    
        
```

```python
dogg=La("ola",15,"yellow","dog")
```

```python
dogg.pr()
```
