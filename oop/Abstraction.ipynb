# Abstraction in Python 

```python
from abc import ABC,abstractmethod
class Shape(ABC):
    @abstractmethod 
    def are(self):
        pass
    def volume(self):
        return "not applicable for given shape"
```

```python
# shape=Shape()
```

```python
class Rectangle(Shape):
    def __init__(self,width,length):
        self.w=width
        self.l=length
    def are(self):
        return self.w*self.l
```

```python
rect= Rectangle(2,3)
```

```python
rect.volume()
```

```python
class cub(Shape):
    def __init__(self,l,w,h):
        self.l=l
        self.w=w
        self.h=h
    def are(self):
        return 6*self.l**2
    def volume(self):
        return self.l**3
```

```python
vo=cub(1,2,3)
```

```python
vo.volume()
```
