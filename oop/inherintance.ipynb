# Single Inherintance

```python
class Animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def into(self):
        print(f"my name is: {self.name}")
    
```

```python
class Cat(Animal):
    def __init__(self,name,age,color):
        super().__init__(name,age)
        self.color=color
    def details(self):
        print(f"{self.name},{self.age}")
    def sound(self):
        print("meoww")
```

```python
cat=Cat("la",10,"black")
```

```python
cat.into()
cat.sound()
```

```python
cat.details()
```

```python
class La(Cat):
    def __init__(self,name,age,color,anni):
        super().__init__(name,age,color)
        self.anni=anni
    def pr(self):
        print(self.name)
        print(self.age)
        print(self.color)
        print(self.anni)
    
        
```

```python
dogg=La("ola",15,"yellow","dog")
```

```python
dogg.pr()
```

# Multiple Inherintance 
one child and more parents

```python
class Parent1:
    def func1(self):
        print("this is parent 1")

class Parent2:
    def func2(self):
        print("this is parent 2")
class Parent3:
    def func1(self):
        print("this is parent 3")
```

```python
class child(Parent1,Parent2,Parent3):
    def func4(self):
        print("this is child function")
```

```python
obj=child()
obj.func2()
```

```python
obj.func1()
```

# Multi Level

```python
class A:
    def method_a(self):
        print("this is method from class A")
class B(A):
    def method_b(self):
        print("this is method from class B")
class C(B):
    def method_c(self):
        print("this is method from class C")
        
```

```python
obj=C()
```

```python
obj.method_a()
obj.method_b()
obj.method_c()
```

```python
class Personn:
    def __init__(self,name:str,age:int):
        self.name=name
        self.age=age
class Emp(Personn):
    def __init__(self,name,age,eid):
        super().__init__(name,age)
        self.eid=eid
class Manag(Emp):
    def __init__(self,name,age,eid,dep):
        super().__init__(name,age,eid)
        self.dep=dep
    def get_details(self):
        print(f"THE NAME IS :{self.name}  AGE :{self.age}  ID :{self.eid} OF DEPARTMENT:{self.dep}")
    
```

```python
laaa=Manag("abi",18,492,"AI/ML")
```

```python
laaa.get_details()
```

# Hierarchical Inheritance

```python
class Animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def intro(self):
        print(f"myname is {self.name}")
        
```

```python
class Cat(Animal):
    def speak(self):
        return "meoww"
class Dog(Animal):
    def speak(self):
        return "woof"
class Lion(Animal):
    def speak(self):
        return "roar"
```
