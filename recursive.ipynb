# Recursive function

```python
#febonacci series
def feb():
    a=1
    b=0
    c=0
    n=int(input("enter the number of terms"))
    print(b)
    for i in range(n):
        c=a+b
        a=b
        b=c
        print(c)
    return
```

```python
feb()
```

```python
def fib(n):
    #base case
    if n==0:
        return 0
    elif n==1:
        return 1
    #recursive case
    else:
        return fib(n-1) + fib(n-2)
```

```python
%%time
fib(5)
```

# Memonization

```python
cache={0:0 , 1:1}
def fibb(n):
    if n in cache:
        return cache[n]
    else:
        re= fibb(n-1) + fibb(n-2)
        cache[n]=re
        return cache[n]
        
```

```python
fibb(70)
cache.clear()
```
