# Arguments in loop


```python
person={
    "name":"abi" ,"surname":"khadka"
,"age":18 }
print(person.items())
print("\n")
for pp in person.items():
     print(pp)
```

```python
for (key,value) in person.items():
    print(key)
    
```

# break and continue

```python
for i in range(1,5):
    if (i==3):
        break
    print(i)
```

```python
for i in range(1,5):
    if(i==2):
        continue
    print(i)
```

```python
#prime number 
u=int(input("enter the number"))
for i in range(u):
    if (u%1==0 and u%u==0):
        print(f"{u} is prime number")
        break
    else:
        continue
```
